```python
linkied list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def detect_and_remove_cycle(head):
    slow = fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            break
    else:
        return False

    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    fast.next = None
    return True


def create_linked_list_with_cycle():
    head = Node(1)
    head.next = Node(2)
    head.next.next = Node(3)
    head.next.next.next = Node(4)

    head.next.next.next.next = head.next

    return head


head = create_linked_list_with_cycle()

if detect_and_remove_cycle(head):
    print("Cycle detected and removed")
else:
    print("No Cycle")
```

```python
def is_balanced(expression):
    stack = []

    for ch in expression:
        if ch in "({[":
            stack.append(ch)

        elif ch in ")}]":
            if not stack:
                return False

            top = stack.pop()

            if ((top == '(' and ch != ')') or
                (top == '{' and ch != '}') or
                (top == '[' and ch != ']')):
                return False

    return len(stack) == 0


expression = input("Enter Expression: ")

if is_balanced(expression):
    print("Balanced Parenthesis")
else:
    print("Not a Balanced Parenthesis")
```

```python
# Program 4: Circular Queue Implementation

class CircularQueue:
    def __init__(self, size):
        self.queue = [None] * size
        self.size = size
        self.front = self.rear = -1

    def enqueue(self, data):
        if (self.rear + 1) % self.size == self.front:
            raise Exception("Queue Overflow")

        if self.front == -1:
            self.front = 0

        self.rear = (self.rear + 1) % self.size
        self.queue[self.rear] = data

    def dequeue(self):
        if self.front == -1:
            raise Exception("Queue Underflow")

        data = self.queue[self.front]

        if self.front == self.rear:
            self.front = self.rear = -1
        else:
            self.front = (self.front + 1) % self.size

        return data

    def display(self):
        if self.front == -1:
            print("Queue Empty")
            return

        print("Queue Elements:", end=" ")
        i = self.front

        while True:
            print(self.queue[i], end=" ")
            if i == self.rear:
                break
            i = (i + 1) % self.size
        print()


cq = CircularQueue(5)

cq.enqueue(10)
cq.enqueue(20)
cq.enqueue(30)
cq.enqueue(40)
cq.enqueue(50)

cq.display()

print("Dequeued:", cq.dequeue())

cq.display()
```

```python
# 4. Circular Queue

class CircularQueue:

    def __init__(self, size):
        self.size = size
        self.queue = []

    def enqueue(self, value):
        if len(self.queue) == self.size:
            print("Queue Full")
        else:
            self.queue.append(value)

    def dequeue(self):
        if len(self.queue) == 0:
            print("Queue Empty")
        else:
            return self.queue.pop(0)


q = CircularQueue(3)

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

print("Deleted:", q.dequeue())
print(q.queue)
```

```python
# Program 5: Student Record using Hashing

class StudentHash:
    def __init__(self):
        self.records = {}

    def insert(self, sid, name, marks):
        self.records[sid] = (name, marks)

    def search(self, sid):
        return self.records.get(sid)

h = StudentHash()

h.insert(101, "Jhon", 85)
h.insert(102, "Max", 90)
h.insert(103, "Zuzi", 78)

sid = int(input("Enter Student ID: "))
result = h.search(sid)

if result:
    print("Student Record Found")
    print("ID:", sid)
    print("Name:", result[0])
    print("Marks:", result[1])
else:
    print("Student Record Not Found")
```

```python
# Program: Binary Search

def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

arr = [10, 20, 30, 40, 50, 60]
target = int(input("Enter element to search: "))

result = binary_search(arr, target)

if result != -1:
    print("Element found at index", result)
else:
    print("Element not found")
```

```python
# Program 10: Recursion - Factorial

def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

n = int(input("Enter a number: "))

if n < 0:
    print("Factorial not defined")
else:
    print("Factorial =", factorial(n))
```

```python
def reverse_string(text):
    stack = []

    for ch in text:
        stack.append(ch)

    reversed_string = ""

    while stack:
        reversed_string += stack.pop()

    return reversed_string


string = input("Enter a string: ")
result = reverse_string(string)

print("Reversed String:", result)
```

```python
# Program 6
# Binary Search Tree: Write a program to find the height of a Binary Search Tree.

class BSTNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


def insert(root, key):
    if root is None:
        return BSTNode(key)

    if key < root.key:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)

    return root


def height(root):
    if root is None:
        return -1

    return 1 + max(height(root.left), height(root.right))


if __name__ == "__main__":
    root = None

    for k in [10, 5, 20, 3, 7]:
        root = insert(root, k)

    print("Calculated Tree Height:", height(root))
```

```python
# Program 7
# Graph Traversal: Implement Depth First Search (DFS) algorithm
# to traverse a given graph.

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)
    print(start, end=" ")

    for neighbour in graph.get(start, []):
        if neighbour not in visited:
            dfs(graph, neighbour, visited)


if __name__ == "__main__":
    graph = {
        'A': ['B', 'C'],
        'B': ['D'],
        'C': [],
        'D': []
    }

    print("DFS Path Sequence:")
    dfs(graph, 'A')
    print()
```

```python
# Program 8
# Sorting Algorithm: Write a program to implement the
# Bubble Sort algorithm and count the number of swaps performed.

def bubble_sort(arr):
    n = len(arr)
    swaps = 0

    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1

    return arr, swaps


if __name__ == "__main__":
    arr = [5, 1, 4, 2, 8]

    sorted_arr, swap_count = bubble_sort(arr)

    print("Sorted Array:", sorted_arr)
    print("Number of Swaps:", swap_count)
```

```python
# Program 12
# Memory Management: Implement a program to merge two sorted arrays
# into a single sorted array with minimal extra space.

def merge_sorted_arrays(arr1, arr2):
    i, j, k = len(arr1) - 1, len(arr2) - 1, len(arr1) + len(arr2) - 1

    arr1.extend([0] * len(arr2))

    while j >= 0:
        if i >= 0 and arr1[i] > arr2[j]:
            arr1[k] = arr1[i]
            i -= 1
        else:
            arr1[k] = arr2[j]
            j -= 1
        k -= 1

    return arr1


if __name__ == "__main__":
    arr1 = [1, 3, 5]
    arr2 = [2, 4, 6]

    print("Merged Output Array:", merge_sorted_arrays(arr1, arr2))
```
