```python


def second_largest(arr):
    if len(arr) < 2:
        return "Invalid Input"

    if arr[0] > arr[1]:
        max1 = arr[0]
        max2 = arr[1]
    else:
        max1 = arr[1]
        max2 = arr[0]

    for i in range(2, len(arr)):
        num = arr[i]

        if num > max1:
            max2 = max1
            max1 = num

        elif num > max2 and num != max1:
            max2 = num

    return max2

arr = [10, 30, 50, 15, 45, 20]
print("Second largest:", second_largest(arr))
```

```python
linkied list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def detect_and_remove_cycle(head):
    slow = fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            break
    else:
        return False

    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    fast.next = None
    return True


def create_linked_list_with_cycle():
    head = Node(1)
    head.next = Node(2)
    head.next.next = Node(3)
    head.next.next.next = Node(4)

    head.next.next.next.next = head.next

    return head


head = create_linked_list_with_cycle()

if detect_and_remove_cycle(head):
    print("Cycle detected and removed")
else:
    print("No Cycle")
```

```python
def is_balanced(expression):
    stack = []

    for ch in expression:
        if ch in "({[":
            stack.append(ch)

        elif ch in ")}]":
            if not stack:
                return False

            top = stack.pop()

            if ((top == '(' and ch != ')') or
                (top == '{' and ch != '}') or
                (top == '[' and ch != ']')):
                return False

    return len(stack) == 0


expression = input("Enter Expression: ")

if is_balanced(expression):
    print("Balanced Parenthesis")
else:
    print("Not a Balanced Parenthesis")
```

```python
# Program 4: Circular Queue Implementation

class CircularQueue:
    def __init__(self, size):
        self.queue = [None] * size
        self.size = size
        self.front = self.rear = -1

    def enqueue(self, data):
        if (self.rear + 1) % self.size == self.front:
            raise Exception("Queue Overflow")

        if self.front == -1:
            self.front = 0

        self.rear = (self.rear + 1) % self.size
        self.queue[self.rear] = data

    def dequeue(self):
        if self.front == -1:
            raise Exception("Queue Underflow")

        data = self.queue[self.front]

        if self.front == self.rear:
            self.front = self.rear = -1
        else:
            self.front = (self.front + 1) % self.size

        return data

    def display(self):
        if self.front == -1:
            print("Queue Empty")
            return

        print("Queue Elements:", end=" ")
        i = self.front

        while True:
            print(self.queue[i], end=" ")
            if i == self.rear:
                break
            i = (i + 1) % self.size
        print()


cq = CircularQueue(5)

cq.enqueue(10)
cq.enqueue(20)
cq.enqueue(30)
cq.enqueue(40)
cq.enqueue(50)

cq.display()

print("Dequeued:", cq.dequeue())

cq.display()
```

```python
# Program 5: Student Record using Hashing

class StudentHash:
    def __init__(self):
        self.records = {}

    def insert(self, sid, name, marks):
        self.records[sid] = (name, marks)

    def search(self, sid):
        return self.records.get(sid)

h = StudentHash()

h.insert(101, "Jhon", 85)
h.insert(102, "Max", 90)
h.insert(103, "Zuzi", 78)

sid = int(input("Enter Student ID: "))
result = h.search(sid)

if result:
    print("Student Record Found")
    print("ID:", sid)
    print("Name:", result[0])
    print("Marks:", result[1])
else:
    print("Student Record Not Found")
```

```python
# Program: Binary Search

def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

arr = [10, 20, 30, 40, 50, 60]
target = int(input("Enter element to search: "))

result = binary_search(arr, target)

if result != -1:
    print("Element found at index", result)
else:
    print("Element not found")
```

```python
# Program 10: Recursion - Factorial

def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

n = int(input("Enter a number: "))

if n < 0:
    print("Factorial not defined")
else:
    print("Factorial =", factorial(n))
```

```python
def reverse_string(text):
    stack = []

    for ch in text:
        stack.append(ch)

    reversed_string = ""

    while stack:
        reversed_string += stack.pop()

    return reversed_string


string = input("Enter a string: ")
result = reverse_string(string)

print("Reversed String:", result)
```
